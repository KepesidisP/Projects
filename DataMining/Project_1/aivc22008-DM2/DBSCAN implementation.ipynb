```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
```

```python
def showPlot(X,IDX):
    
    plt.figure(1)
    plt.scatter(X[:,0],X[:,1])
    plt.show()
    
    plt.figure(2)
    plt.scatter(X[:,0],X[:,1],c=IDX)
    plt.show()
    
```

# mydata Dataset

```python
#mydata.mat

mat_file= scipy.io.loadmat('mydata.mat')
X=np.array(mat_file['X'])

epsilon=0.5
MinPts=15
dbscan= DBSCAN(epsilon, min_samples=MinPts).fit(X)
IDX= dbscan.labels_

showPlot(X,IDX)
```

# Iris Dataset

```python
#Iris data

meas=load_iris().data
X=meas[:,[2,3]]

epsilon=0.1
MinPts=5
dbscan= DBSCAN(epsilon,min_samples=MinPts).fit(X)
IDX= dbscan.labels_
showPlot(X,IDX)
```

```python
# 'Βήμα 5'
xV1=zscore(X[:,0])
xV2=zscore(X[:,1])
for i in range(len(xV1)):
    X[i][0]=xV1[i]
    X[i][1]=xV2[i]

epsilon=0.1
MinPts=5
dbscan= DBSCAN(epsilon,min_samples=MinPts).fit(X)
IDX= dbscan.labels_

showPlot(X,IDX)
```

# xV Dataset

```python
#xV dataset

mat_file= scipy.io.loadmat('xV.mat')
xV=np.array(mat_file['xV'])

X= xV[:,[0,1]]
epsilon=0.3
MinPts=50
dbscan= DBSCAN(epsilon,min_samples=MinPts).fit(X)
IDX= dbscan.labels_

showPlot(X,IDX)
```
